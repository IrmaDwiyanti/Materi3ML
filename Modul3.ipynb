````python
# reading the database
data = pd.read_csv("/content/drive/MyDrive/SEM6/Prak. Pembelajaran Mesin/materi3/students.csv")

# printing the top 10 rows
display(data.head(10))
````

````python
# Show the dataset info
data.info()
````

````python
# Missing Value Checking
data.isna().sum()
````

````python
# Filling Nan Value with mode
data['lunch'].fillna(data['lunch'].mode()[0], inplace=True)
````

````python
# Filling Nan Value with mean
data.loc[:, 'reading score'] = data['reading score'].fillna(data['reading score'].mean())
````

````python
# Filling Nan Value with mode
data.loc[:, 'grade'] = data['grade'].fillna(data['grade'].median())
````

````python
# Check the information of the dataset
data.info()
````

**Perhatikan**

Alternatif lain selain handling missing value dengan nilai mean, median atau modus, kita juga dapat melakukan handling missing value dengan teknik droping (menghapus baris atau kolom dengan fungsi drop()), interpolasi, backwardfill ataupun forwardfill seperti dibawah ini.



*   Menggunakan teknik Interpolasi, teknik ini adalah teknik estimasi nilai yang hilang berdasarkan pola atau tren dari data yang ada. Teknik ini umum digunakan dalam data yang bentuknya  time series atau data numerik dengan hubungan kontinu. Teknik Interpolasi ini dapat dilakukan seperti dibawah ini.

Linear Interpolation Menghitung nilai yang hilang dengan membuat garis lurus antara dua titik data yang diketahui.




```
# Interpolate the Missing Value
data['nama_fitur'].interpolate(method='linear', inplace=True)
```




*   Menggunakan teknik backwardfill atau forwardfill, teknik backwardfill adalalh teknik handling missing value yang akan mengisi nilai Nan atau nilai kosong dengan nilai setelahnya sedangkan untuk forwardfill kebalikannya, teknik ini akan mengisi nilai Nan dengan nilai setelahnya contohnya seperti dibawah ini.



```
#backwardfill the missing value
data['nama_fitur'].fillna(method='bfill', inplace=True)
```


```
#forwardfill the missing value
data[‘’nama_fitur].filna(method= ‘bfill’, inplace = True)

```


*   Menggunakan teknik dropping, teknik ini perlu digunakan secara hati - hati dan penuh pertimbangan dikarenakan teknik ini akan menghapus data baris atau kolom Nan kita secara permanen, teknik ini umum digunakan jika data dengan jumlah Nan hanya sedikit ataupun data dengan jumlah Nan lebih banyak dari 50% data kita. Contohnya seperti dibawah ini.



```
# drop Nan Value
data.dropna(subset=['nama_fitur'], inplace=True)

Atau

# drop feature with more than 50% are Nan value
data.drop(columns=['nama_fitur'], inplace=True)
```
